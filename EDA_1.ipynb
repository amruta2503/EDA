```python
col_int = data.select_dtypes(include = 'int').columns.tolist()
print("col_int\n",col_int)
```

```python
col_obj = data.select_dtypes(include = 'O').columns.tolist()
print("col_obj\n",col_obj)
```

```python
for feature in col_int:
    data.groupby('gender')[feature].median().plot.bar()
    plt.xlabel("gender")
    plt.ylabel(feature)
    plt.show()
```

```python
# ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course', 'Result']

for feature in col_int:
    data.groupby('race/ethnicity')[col_int].median().plot.bar()
    plt.xlabel('race/ethnicity')
    plt.ylabel(feature)
    plt.show
```
